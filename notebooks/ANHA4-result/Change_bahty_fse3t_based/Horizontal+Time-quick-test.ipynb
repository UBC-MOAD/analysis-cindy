```python
import matplotlib.cm as cm
def vetical(data,dep,string):
    a=np.where(data==data.max())
    print a
    for i in range (len(data[:,1,1,1])):
        cc=cm.Spectral(i*25)
        plt.plot(data[-i,:,a[2][0],a[3][0]],dep,color=cc,linewidth=1,label='T:%s'%i)
        plt.legend()

    for k in range (len(dep)):
        plt.axhline(dep[k],c="grey",linestyle='--',linewidth=0.5,alpha=0.5)
    plt.axhline(bathy[a[2][0],a[3][0]],c="grey",linewidth=5,alpha=0.7)
    plt.ylabel('depth (m)')
    plt.xlabel('%s Conc.'%string)
    plt.gca().invert_yaxis()
    print ('deepest:%s m'%bathy[a[2][0],a[3][0]])
```

```python
def plot_inf(filenum,num,string,grey):
    ptrc=nc.Dataset(nc_filename[filenum])
    print 'filename:%s'%nc_filename[filenum]
    dep=ptrc.variables['deptht'][:]
    if string =='time':
        if num==7:
            TR_7=ptrc.variables['TR_7'][:,42,:,:]
            TR_7=np.ma.masked_where(TR_7<= 0, TR_7)
            trc=np.ma.masked_where(np.isnan(TR_7),TR_7)
            critialnum=0.6
        if num==8:
            TR_8=ptrc.variables['TR_8'][:,42,:,:]
            TR_8=np.ma.masked_where(TR_8<= 0, TR_8)
            trc=np.ma.masked_where(np.isnan(TR_8),TR_8)
            critialnum=1
    else:
        if num==7:
            TR_7=ptrc.variables['TR_7'][-1,40:48,:,:]
            TR_7=np.ma.masked_where(TR_7<= 0, TR_7)
            trc=np.ma.masked_where(np.isnan(TR_7),TR_7)
            critialnum=0.6
        if num==8:
            TR_8=ptrc.variables['TR_8'][-1,40:48,:,:]
            TR_8=np.ma.masked_where(TR_8<= 0, TR_8)
            trc=np.ma.masked_where(np.isnan(TR_8),TR_8)
            critialnum=1

    data=trc
    fig = plt.figure(figsize=(20,15))
    for k in range (len(data[:,0,0])):
        ii=[]
        jj=[]
        for i in range (len(data[0,:,0])):
            for j in range (len(data[0,0,:])):
                if data[k,i,j]>critialnum:
                    ii.append(i)
                    jj.append(j)

        ax = fig.add_subplot(3,4,k+1)
        #if k==0:
        plt.pcolor(data[k,:,:])
        #else:
        #    plt.pcolor(data[k,:,:]-data[k-1,:,:])
        plt.colorbar()
        if grey=='yes':
            plt.scatter(jj,ii,alpha=0.3,color='grey',s=50)
        if string =='time':
            plt.title('t=%sm, # of unusual points=%s'%(k+0.5,len(jj)))
            #plt.ylim([450,800])
        else:
            plt.title('timestep=-1,dep=%s m'%(dep[40+k]))
        #plt.clim([0.0,0.2])        
    plt.show()
```

```python
def verticalplot(lat,data1, title1,data2,title2,lmin=30,lmax=46,ylim=[0,4000]):
    fig = plt.figure(figsize=(20,5))
    ax = fig.add_subplot(1,2,1)
    plt.contourf(lat[:],dep[lmin:lmax],data1[lmin:lmax,:],alpha=0.5)
    plt.colorbar()
    plt.ylim(ylim)
    plt.gca().invert_yaxis()
    plt.ylabel('Depth (m)')
    plt.xlabel('Lat')
    plt.title('%s'%title1)
    plt.axhline(y=1062,c="blue",linewidth=2,alpha=0.5)
    plt.axhline(y=2225,c="blue",linewidth=2,alpha=0.5)
    plt.axhline(y=3220,c="blue",linewidth=2,alpha=0.5)

    ax = fig.add_subplot(1,2,2)
    plt.contourf(lat[:],dep[lmin:lmax],data2[lmin:lmax,:],alpha=0.5)
    plt.colorbar()
    plt.ylim(ylim)
    plt.gca().invert_yaxis()
    plt.ylabel('Depth (m)')
    plt.xlabel('Lat')
    plt.title('%s'%title2)
    plt.axhline(y=1062,c="blue",linewidth=2,alpha=0.5)
    plt.axhline(y=2225,c="blue",linewidth=2,alpha=0.5)
    plt.axhline(y=3220,c="blue",linewidth=2,alpha=0.5)

    plt.show()
```
