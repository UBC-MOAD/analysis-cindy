```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## High Ice

```python
# CESAR 1983
depth = [0, 120, 200, 400, 700, 1200, 1500]
dissolved = [0.26, 0.19, 0.51, 0.62, 1.03, 1.1, 1.14]
particulate = [float('NaN'), 0.012, 0.027, 0.17, 0.078, 0.055, 0.11]
# 1991 Arctic Expedition Stn 3
depth = depth +[50, 500, 1000, 1850, 2850, 3850]
dissolved = dissolved + [0.021, 0.574, 0.684, 1.085, 1.331, 1.385]
particulate = particulate + [0.02, 0.057, 0.088, 0.159, 0.188, 0.212]

pd = np.polyfit((depth[3:7]+depth[8:]), (dissolved[3:7]+dissolved[8:]), 1)
print(depth[3:7]+depth[8:])
pp = np.polyfit(depth[3:7]+depth[8:], particulate[3:7]+particulate[8:], 1)
drange = np.arange(0, 4000, 50)

fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].plot(np.array(particulate), np.array(depth), 'o')
ax[0].plot(pp[1] + pp[0]*drange, drange)
ax[0].invert_yaxis()
ax[1].plot(dissolved, depth, 'o')
ax[1].plot(pd[1] + pd[0]*drange, drange)
ax[1].invert_yaxis()
```

```python
Q = 0.0267
# Particulate Slope in v High ice
mP = pp[0]
S = Q/mP
print (S)
```

```python
mD = pd[0]
kdOka = mD/mP
print (kdOka)
```

```python
# if kd = 1.6
ka = 1.6/kdOka
print (ka)
```

```python
P0 = pp[0]*500 + pp[1] 
print (Q/ka + kdOka*P0)
print (pd[0]*500 + pd[1])
```

```python
# if kd = 1.0
ka = 1.0/kdOka
print (ka)
```

```python
print (Q/ka + kdOka*P0)
difference = Q/ka + kdOka*P0 - (pd[0]*500 + pd[1])
print (difference)
P0f = P0 - difference/2/kdOka
```

```python
# if kd = 0.5
ka = 0.5/kdOka
print (ka)
```

```python
print (Q/ka + kdOka*P0)
difference = Q/ka + kdOka*P0 - (pd[0]*500 + pd[1])
print (difference)
P0f = P0 - difference/2/kdOka
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].plot(np.array(particulate), np.array(depth), 'o')
ax[0].plot(pp[1] + pp[0]*drange, drange)
ax[0].invert_yaxis()
ax[1].plot(dissolved, depth, 'o')
ax[1].plot(pd[1] + pd[0]*drange, drange)
ax[1].invert_yaxis()
ax[0].plot(P0f + Q/S*(drange-500), drange)
ax[1].plot(Q/ka + P0f*kdOka + kdOka*Q/S*(drange-500), drange)
print (Q/((pp[1] - difference/2/kdOka)/500.))
```

```python
import statsmodels.api as sm
ice=[0.923636,0.35]   #NASA
k=[0.0852111333524,0.164232559931]
s=[715.287907601,994.100601806]
```

```python
ice_ = sm.add_constant(ice,prepend=False)

paraS=sm.WLS(s,ice_).fit()
paraK=sm.WLS(k,ice_).fit()
```

```python
# s and ice
x=np.array([0,1])
yS=x*paraS.params[0]+paraS.params[1]
print x,yS
plt.plot(x,yS)
plt.plot(np.array(ice), np.array(s), 'o')
```

```python
#k and ice
yK=x*paraK.params[0]+paraK.params[1]
print x,yK
plt.plot(x,yK)
plt.plot(np.array(ice), np.array(k), 'o')
```
