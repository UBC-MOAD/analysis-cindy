# Fit parameters 

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## High Ice

```python
# 1991 Arctic Expedition Stn 3
depth = [50, 500, 1000, 1850, 2850, 3850]
dissolved = [0.062, 0.216, 0.258, 0.539, 0.632, 0.622]
particulate = [0.007, 0.006, 0.007, 0.01, 0.013, 0.012]

pd = np.polyfit(depth[1:], dissolved[1:], 1)
pp = np.polyfit(depth[1:], particulate[1:], 1)
drange = np.arange(0, 4000, 50)

fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].plot(np.array(particulate), np.array(depth), 'o')
ax[0].plot(pp[1] + pp[0]*drange, drange)
ax[0].invert_yaxis()
ax[1].plot(dissolved, depth, 'o')
ax[1].plot(pd[1] + pd[0]*drange, drange)
ax[1].invert_yaxis()
```

```python
Q = 0.00246
# Particulate Slope in v High ice
mP = pp[0]
S = Q/mP
print (S)
```

```python
mD = pd[0]
kdOka = mD/mP
print (kdOka)
```

```python
'''input kd'''
kd=0.3
```

```python
# if kd = 1.0
ka = kd/kdOka
print (ka)
```

```python
# test result
P0 = pp[0]*500 + pp[1] 
print (Q/ka + kdOka*P0)
print (pd[0]*500 + pd[1])
```

```python
print (Q/ka + kdOka*P0)
difference = Q/ka + kdOka*P0 - (pd[0]*500 + pd[1])
print (difference)
P0f = P0 - difference/2/kdOka
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].plot(np.array(particulate), np.array(depth), 'o')
ax[0].plot(pp[1] + pp[0]*drange, drange)
ax[0].invert_yaxis()

ax[1].plot(dissolved, depth, 'o')
ax[1].plot(pd[1] + pd[0]*drange, drange)
ax[1].invert_yaxis()

ax[0].plot(P0f + Q/S*(drange-500), drange)
ax[1].plot(Q/ka + P0f*kdOka + kdOka*Q/S*(drange-500), drange)
print (Q/((pp[1] - difference/2/kdOka)/500.))
```
