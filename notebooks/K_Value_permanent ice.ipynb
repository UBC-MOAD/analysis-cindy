# Equation: [X]diss= Q/K-J/K*(Q/s*z+C)

[X]diss: concentration of dissolved Pa (or Th)
Q: production rate of X(Pa or Th)
K: adsorbtion rate of X
J: desorption rate of X
s: sinking rate (500m/y)
C: just a constant

```python
Q_Th=0.0267
s_Th=500
c_Th=-0.025
c_Th2=0.025

k_Th=0.15      #absorption rate
j_Th=1.6    #desorption rate
z = np.linspace(-4000,0,100,endpoint=True)
x_Th = np.array(Q_Th/k_Th-j_Th/k_Th*(Q_Th/s_Th*z+c_Th))
x_Th2 = np.array(Q_Th/k_Th-j_Th/k_Th*(Q_Th/s_Th*z+c_Th2))

fig = plt.figure(1, figsize=(18,6)) 
#fig1
plt.subplot(1, 2, 1)
plot (x_Th, -z, color='blue', alpha=0.30)
plot (x_Th2, -z, color='red', alpha=0.30)

plt.scatter(th,dep)
plt.scatter(th4,dep4,color='lightyellow')
plt.scatter(th2,dep2, color='red')
plt.scatter(th3,dep3,color='orange')
plt.xlabel('Water-column diss_Th  (dpm/1000litres)')
plt.ylabel('Depth (m)')
plt.title("Concentration of Dissolved Th")

plt.axis([-2,2,0,4000])
plt.gca().invert_yaxis()

#fig2
fig = plt.figure(1, figsize=(15,5)) 
#fig1
plt.subplot(1, 2, 2)
z = np.linspace(-4000,0,100,endpoint=True)
x_Th = np.array(Q_Th/k_Th-j_Th/k_Th*(Q_Th/s_Th*z+c_Th))
x_Th2 = np.array(Q_Th/k_Th-j_Th/k_Th*(Q_Th/s_Th*z+c_Th2))
plot (x_Th, -z, color='blue', alpha=0.30,label="surface: K(adsorbtion rate)=0.15,J(desorption rate)=1.6")
plot (x_Th2, -z, color='red', alpha=0.30)

z = np.linspace(-4000,-750,100,endpoint=True)
k_Th=1.55      #absorption rate
j_Th=1.6    #desorption rate
c_Th=-0.95
c_Th2=-0.3
x_Th = np.array(Q_Th/k_Th-j_Th/k_Th*(Q_Th/s_Th*z+c_Th))
x_Th2 = np.array(Q_Th/k_Th-j_Th/k_Th*(Q_Th/s_Th*z+c_Th2))

plot (x_Th, -z, color='darkblue', alpha=0.30,label="deeper: K(adsorbtion rate)=1.55,J(desorption rate)=1.6")
plot (x_Th2, -z, color='darkred', alpha=0.30)
plt.scatter(th,dep)
#plt.scatter(th4,dep4,color='yellow')
plt.scatter(th2,dep2, color='red')
plt.scatter(th3,dep3,color='orange')
plt.axis([-2,2,0,4000])
plt.legend(loc=3,fontsize=11)
plt.xlabel('Water-column diss_Th  (dpm/1000litres)')
plt.ylabel('Depth (m)')
plt.gca().invert_yaxis()
plt.title("Concentration of Dissolved Th")

show()
```
