# Th on AR

```python

def ana(xrange=4000,i=0.9):
    x=np.linspace(0,xrange,20)
    
    pth=0.0267/(-1115*i+1600)*x+0.03
    k  =-0.12*i+0.21
    dth=0.0267/k+0.5/k*pth

    ppa=0.00246/(-508*i+1616)*x+0.004
    k_pa=-0.055*i+0.082
    dpa=0.00246/k_pa+1.5/k_pa*ppa

    totalth=pth+dth
    totalpa=ppa+dpa
    return x,totalth, totalpa
```

```python
def ana_realice(x,i):

    pth=0.0267/(-1115*i+1600)*x+0.03
    k  =-0.12*i+0.21
    dth=0.0267/k+0.5/k*pth

    ppa=0.00246/(-508*i+1616)*x+0.004
    k_pa=-0.055*i+0.082
    dpa=0.00246/k_pa+1.5/k_pa*ppa

    totalth=pth+dth
    totalpa=ppa+dpa
    return x,totalth, totalpa
```
