```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import numpy as np
import copy
import os
import glob
from mpl_toolkits.basemap import cm,addcyclic,Basemap, shiftgrid
% matplotlib inline
```

```python
nc_filename=glob.glob('*.nc')
print("found *.nc file: \n{}".format(nc_filename))
```

```python
W=nc.Dataset(nc_filename[0])
U=nc.Dataset(nc_filename[7])
V=nc.Dataset(nc_filename[5])

ptrc_T=nc.Dataset(nc_filename[2])
print ptrc_T.variables
print "      "
print W.variables
```

```python
lat = ptrc_T.variables['nav_lat'][:]
lon = ptrc_T.variables['nav_lon'][:]
Dis_Pa = ptrc_T.variables['TR_7'][:]
Par_Pa = ptrc_T.variables['TR_9'][:]
Dis_Th = ptrc_T.variables['TR_8'][:]
Par_Th = ptrc_T.variables['TR_10'][:]
depth = ptrc_T.variables['deptht'][:]
print('Latitude {} ~ {}'.format(np.min(lat), np.max(lat)))
print('Longitude {} ~ {}'.format(np.min(lon), np.max(lon)))
```

```python
u=U.variables['vozocrtx'][:]
v=V.variables['vomecrty'][:]
w=W.variables['vovecrtz'][:]
```

```python
def delete_edge(data, order=[1, 1, 1, 1]):
    import numpy as np

    data[:, 0:order[0]]=0
    data[0:order[1], :]=0
    data[:, (np.size(data, 1)-order[2]):np.size(data, 1)]=0
    data[(np.size(data, 0)-order[3]):np.size(data, 0), :]=0

    return data
u_mask=delete_edge(u, [1, 1, 1, 1])
v_mask=delete_edge(v, [1, 1, 1, 1])
w_mask=delete_edge(w, [1, 1, 1, 1])
```

```python
latlim=[10, 50]
lonlim=[-85, -48]
```

```python
Dis_Pa = np.ma.masked_equal(Dis_Pa,0)
Dis_Th = np.ma.masked_equal(Dis_Th,0)
Par_Pa = np.ma.masked_equal(Par_Pa,0)
Par_Th = np.ma.masked_equal(Par_Th,0)
u_mask = np.ma.masked_equal(u_mask,0)
v_mask = np.ma.masked_equal(v_mask,0)
w_mask = np.ma.masked_equal(w_mask,0)
print('Dis_Pa {} ~ {}'.format(np.min(Dis_Pa), np.max(Dis_Pa)))
print('Dis_Th {} ~ {}'.format(np.min(Dis_Th), np.max(Dis_Th)))
print('Par_Pa {} ~ {}'.format(np.min(Par_Pa), np.max(Par_Pa)))
print('Par_Th {} ~ {}'.format(np.min(Par_Th), np.max(Par_Th)))
```
