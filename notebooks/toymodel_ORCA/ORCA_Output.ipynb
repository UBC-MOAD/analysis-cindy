```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import numpy as np
import copy
import os
import glob
from mpl_toolkits.basemap import cm,addcyclic,Basemap, shiftgrid
% matplotlib inline
```

```python
nc_filename=glob.glob('ORCA2_5*.nc')
print("found *.nc file: \n{}".format(nc_filename))
```

```python
T_name=glob.glob('ORCA2_5d_00010101_00011231_grid_T.nc')
U=nc.Dataset(nc_filename[1])
W=nc.Dataset(nc_filename[3])
V=nc.Dataset(nc_filename[4])
T_obj=nc.Dataset(T_name[0])
```

```python
nav_lon=T_obj.variables['nav_lon'][:]
nav_lat=T_obj.variables['nav_lat'][:]
sst=T_obj.variables['sosstsst'][:]
print nav_lat.shape
print sst[0,:,:].shape
print('Latitude {} ~ {}'.format(np.min(nav_lon), np.max(nav_lon)))
print('Longitude {} ~ {}'.format(np.min(nav_lat), np.max(nav_lat)))
```

```python
lonlim=[-180, 180]
latlim=[-78, 89]
```

```python
# create figure and axes instances
fig = plt.figure(figsize=(9,9))
ax = fig.add_axes([0.1,0.1,0.8,0.8])

# create Basemap instance.
m = Basemap(projection='npstere',boundinglat=53,lon_0=0,resolution='l')
m.drawcoastlines()
m.fillcontinents(color = '0.85')
m.drawparallels(np.arange(-80.,81.,15.),labels=[1, 0, 0, 0])
m.drawmeridians(np.arange(-180.,181.,20.),labels=[0, 0, 0, 1])

x, y = m(nav_lon, nav_lat)
clev=np.arange(20, 40, 2.5)
cs = m.contourf(x,y,sst[1,:,:],cmap=plt.cm.RdBu_r)
cbar = m.colorbar(cs,location='right',pad="2%")
plt.title("SST")
plt.show()
```

```python
u=U.variables['vozocrtx'][:]
v=V.variables['vomecrty'][:]
w=W.variables['vovecrtz'][:]
#u_mask = np.ma.masked_equal(u_mask,0)
#v_mask = np.ma.masked_equal(v_mask,0)
#w_mask = np.ma.masked_equal(w_mask,0)
```

```python
fig = plt.figure(figsize=(9,9))
m = Basemap(projection='npstere',boundinglat=70,lon_0=0,resolution='l')
m.drawcoastlines()
m.fillcontinents(color = '0.85')
m.drawparallels(np.arange(-80.,81.,15.),labels=[1, 0, 0, 0])
m.drawmeridians(np.arange(-180.,181.,20.),labels=[0, 0, 0, 1])

x, y = m(nav_lon, nav_lat)
Q=m.quiver(x, y, u[36,5,:,:], v[36,5,:,:],scale=0.6)
#smaller the number, longer the line
# make quiver key.
Qlabel=plt.quiverkey(Q, 0.15, 0.1, 0.03, '0.03 m/s', labelpos='N') # width, height, scale, label, position
ax.set_title('NEMO_ORCA_surface_currents', fontsize=12)
```

```python
fig = plt.figure(figsize=(9,9))
#m =Basemap(llcrnrlon=0,llcrnrlat=-90,urcrnrlon=360,urcrnrlat=90,projection='mill')
m1 = Basemap(resolution='l',projection='nplaea',boundinglat=70,lon_0=0)


m1.drawcoastlines(linewidth=1.25)
m1.fillcontinents(color='0.8')
m1.drawparallels(np.arange(-80,81,20),labels=[1,1,0,0])
m1.drawmeridians(np.arange(0,360,60),labels=[0,0,0,1])

x, y = m1(nav_lon, nav_lat)
Q=m1.quiver(x, y, u[36,5,:,:],v[36,5,:,:], width=0.003, scale=0.6)

plt.title('current-North-Polar Lambert Azimuthal')
plt.show()
```
